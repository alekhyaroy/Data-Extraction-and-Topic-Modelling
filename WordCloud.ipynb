```python
#Importing libraries

import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import csv
```

```python
file_ob = open(r"clean_excel_apple.csv", encoding = "ISO-8859-1")
```

```python
# reader object is created
reader_ob = csv.reader(file_ob)
# contents of reader object is stored .
# data is stored in list of list format.
reader_contents = list(reader_ob)
```

```python
# empty string is declared
text = ""
```

```python
# iterating through list of rows
for row in reader_contents :
      
    # iterating through words in the row
    for word in row :
  
        # concatenate the words
        text = text + " " + word
```

```python
stopwords = set(STOPWORDS)
```

```python
# show only 'n' words in the wordcloud .
wordcloud = WordCloud(width=1500, height=1500, max_words=100, random_state=100, stopwords=stopwords, ).generate(text)
```

```python
# plot the WordCloud image 
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
